```python
import pickle
import numpy as np

# 1) Charger le modèle
with open("model.pickle", "rb") as f:
    model = pickle.load(f)

# 2) Vérifier s’il existe une méthode `partial_fit`
has_partial_fit = hasattr(model, "partial_fit")

print("Ce modèle a-t-il une méthode partial_fit ? :", has_partial_fit)

if has_partial_fit:
    print("=> Le modèle est potentiellement incrémental.")

    # 3) Facultatif : tester un mini-apprentissage incrémental
    #    On crée un jeu de données bidon juste pour la démo :
    X_dummy = np.random.rand(10, 1024)  # par ex. 10 échantillons, 1024 features
    y_dummy = np.random.randint(0, 2, 10)  # étiquettes binaires (0 ou 1)
    
    try:
        # Si c'est un classifieur scikit-learn classique, on peut
        # donner `classes=np.unique(y_dummy)` ou la liste complète des classes
        model.partial_fit(X_dummy, y_dummy, classes=np.unique(y_dummy))
        print("La méthode partial_fit a fonctionné sans erreur : le modèle est bien incrémental.")
    except Exception as e:
        print("Erreur lors de l'appel à partial_fit :", e)
else:
    print("=> Le modèle ne dispose pas de partial_fit, donc il n'est pas incrémental.")
```
